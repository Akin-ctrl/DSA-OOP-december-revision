 Implement your own Stack: push(x), pop(), peek(), is_empty().
 
 Use it to: Check if parentheses string "()[]{}" is valid: Function is_valid_parentheses(s).

```python
class Stack():
    def __init__(self):
        self.items = []

    def push(self, item):
        '''Push element x onto stack.'''
        self.items.append(item)

    def pop(self):
        '''Remove the top element from stack and return it.'''
        if not self.is_empty():
           value = self.items[-1]
           del self.items[-1]
           return value
        return None

    def peek(self):
        '''Get the top element.'''
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        '''Return whether the stack is empty.'''
        return len(self.items) == 0





```

```python
class ParenthesesValidator():
    def __init__(self):
        self.parentheses_map = {')': '(', '}': '{', ']': '['}

    def is_valid(self, s: str) -> bool:
        stack = Stack()  # Create fresh stack for each validation
        for char in s:
            if char in self.parentheses_map.values():
                stack.push(char)
            elif char in self.parentheses_map.keys():
                if stack.is_empty():
                    return False 
                top_element = stack.pop()
                if top_element != self.parentheses_map[char]:
                    return False
        return stack.is_empty()


# Test
validator = ParenthesesValidator()
print("'()[]{}' is valid:", validator.is_valid("()[]{}"))  # True
print("'([)]' is valid:", validator.is_valid("([)]"))      # False
print("'{[]}' is valid:", validator.is_valid("{[]}"))      # True
```
