Sliding Window (Intro)

Implement max_sum_subarray(nums, k):

Returns max sum over all contiguous subarrays of size k.
Must use sliding window, not nested loops.


```python
nums = []
def max_sum_subarray(nums, k: int=3):
    if k <= 0:
        raise ValueError("k must be a positive integer")
    if len(nums) < k:
        raise ValueError("The length of nums must be at least k")
    window_sum = sum(nums[:k])
    max_sum = window_sum
    for i in range(k, len(nums)):
        window_sum +=nums[i] - nums[i-k]
        if window_sum > max_sum:
            max_sum = window_sum
            
    return max_sum
```

```python
if __name__ == "__main__":
    k = 4
    while True:
        user_input = input("Enter a number (or 'done' to finish): ")
        if user_input.lower() == 'done' or user_input == '':
            break
        try:
            num = float(user_input)
            nums.append(num)
        except ValueError:
            print("Invalid input. Please enter a number.")
   
   
    output = max_sum_subarray(nums, k)
    print("The max sum of the subarray of size", k, "is:", output)
```

Create WindowedStream class:

Initialized with window_size.

Method .add(value) feeds values.

Method .current_window() returns last window_size items.

Test with sample data.

```python
class WindowedStream():
    def __init__(self, k: int = 3):
        self._window_size = k
        self._values = []
    
    def add(self, value) -> None:
        self._values.append(value)
        if len(self._values) > self._window_size:
            self._values.pop(0)

    def current_window(self) -> list:
        return self._values
    
```

```python
if __name__ == "__main__":
    k = 5
    window = WindowedStream(k)
    while True:
        user_input = input("Enter a number (or 'done' to finish): ")
        if user_input.lower() == 'done' or user_input == '':
            break
        try:

            num = float(user_input)
            window.add(num)
        except ValueError:
            print("Invalid input. Please enter a number.")

    if len(window.current_window()) < k:
        print(f"Need at least {k} numbers. You entered {len(window.current_window())}.")
    else:
        print("Current window:", window.current_window())
        print("Window sum:", sum(window.current_window()))
```
